### Find the car (or cars) with the best ratio of horsepower to weight among all cars that also have above-median mpg.

```python
#your code here
median_mpg=df['mpg'].median()
avg_median_mpg=df[df['mpg']>median_mpg].copy()
avg_median_mpg['hp_per_weight']= avg_median_mpg['horsepower']/avg_median_mpg['weight']
max= avg_median_mpg['hp_per_weight'].max()

best_cars= avg_median_mpg[avg_median_mpg['hp_per_weight']== max]
print(best_cars)


```

### Design a multi-line plot using Matplotlib or Seaborn that shows the evolution of average mpg over the years, separately for each origin

```python
#your code here
avg_mpg=df.groupby(['model_year','origin'])['mpg'].mean().reset_index()

plt.figure(figsize=(10,6))
sns.lineplot(data=avg_mpg,x='model_year',y='mpg',hue='origin',marker='o')
plt.title('average mpg over years for each origin')
plt.xlabel('Model Year')
plt.ylabel('Average mpg')
plt.legend(title='Origin')
plt.grid(True)
plt.tight_layout()

plt.show()
```

### Create a Seaborn scatterplot (or PairGrid) where:

X = horsepower

Y = weight

Color by: origin

Size by: mpg

Hue order = ['japan', 'europe', 'usa']

Add meaningful plot titles and axis titles.

```python
#your code here
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# 1. Drop duplicate columns (if any)
df = df.loc[:, ~df.columns.duplicated()]

# 2. Ensure horsepower, weight, mpg are numeric
for col in ['horsepower', 'weight', 'mpg']:
    df[col] = pd.to_numeric(df[col], errors='coerce')

# 3. Drop rows with missing values in the columns we care about
df = df.dropna(subset=['horsepower', 'weight', 'mpg', 'origin'])

# 4. Clean up 'origin' values
df['origin'] = df['origin'].str.strip().str.lower()

# 5. Create scatter plot
plt.figure(figsize=(10, 6))
sns.scatterplot(
    data=df,
    x='horsepower',
    y='weight',
    hue='origin',
    size='mpg',
    hue_order=['japan', 'europe', 'usa'],
    sizes=(20, 200),
    alpha=0.7,
    palette='Set2'
)

# 6. Labels and formatting
plt.title('Horsepower vs Weight of Cars\n(Size by MPG, Color by Origin)', fontsize=14)
plt.xlabel('Horsepower')
plt.ylabel('Weight')
plt.legend(title='Origin', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.grid(True)
plt.tight_layout()
plt.show()
```

### We define a consistent car model as one that was produced over multiple years and had very low variation in mpg across those years (standard deviation < 1.0).

Tasks:

Identify car names that appear in more than one model_year.

For each such name, compute the standard deviation of mpg across years.

Return the car(s) with the lowest variation in mpg, among those with at least 2 appearances and std(mpg) < 1.0.

Report the model name(s), number of appearances, and the average mpg.

Bonus:
Sort the result by number of appearances (descending), then mpg (descending).

```python
#your code here
grouped=df.groupby('name')
multiyrs_cars=grouped.filter(lambda x:x['model_year'].nunique()>1)

summary= multiyrs_cars.groupby('name').agg(
    count=('mpg', 'count'),
    mpg_std=('mpg', 'std'),
    avg_mpg=('mpg', 'mean')
)

# Filter where count >= 2 and std(mpg) < 1.0
filtered = summary[(summary['count'] >= 2) & (summary['mpg_std'] < 1.0)]

# Final sorting
result = filtered.sort_values(by=['count', 'avg_mpg'], ascending=[False, False])

# Optional: round for display
result = result.round({'mpg_std': 2, 'avg_mpg': 2})

# Display result
print(result.reset_index())
```
